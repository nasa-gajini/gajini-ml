# Aboveground Biomass Density
* Aboveground Biomass (AGB) is a widely-used variable that helps characterize **the growth of crops and forecast their yields.**
* 지상부 바이오매스 밀도(AGB, Aboveground Biomass Density)에 대한 데이터를 요약한 표입니다. AGB는 작물의 성장과 생산량 예측을 돕는 중요한 변수로 사용됩니다. 

- **Spatial Resolution (공간 해상도)**: 25m 직경, 1km 격자.
- **Spatial Coverage (공간 범위)**: 북위 52도에서 남위 52도까지의 위도 범위.
- **Temporal Resolution (시간 해상도)**: 한 번의 추정치.
- **Temporal Coverage (시간 범위)**: 2019년부터 2022년까지.
- **Spectral Resolution (스펙트럼 해상도)**: 레이저 파장의 길이는 1.064 마이크로미터.
- **Satellite/Platform (위성/플랫폼)**: 국제우주정거장(International Space Station).
- **Sensor(s)/Model Name (센서/모델 이름)**: GEDI.
- **Observation or Model (관측 또는 모델)**: 관측 및 모델링.
- **File Format (파일 형식)**: GeoTIFF.

이 데이터는 주로 AGB 추정치를 제공하며, 이를 통해 바이오매스와 탄소 저장량을 분석할 수 있습니다.

# ESI 

* 증발 스트레스 지수(ESI, Evaporative Stress Index)**에 대한 데이터를 요약한 표가 포함되어 있습니다. ESI는 증발산(ET)에서 시간에 따른 이상 현상을 설명하며, 지표면에서 물 사용률이 비정상적으로 높거나 낮은 지역을 강조합니다. 또한, 장기간의 고온, 건조, 바람이 강한 조건에서 발생하는 '플래시 가뭄'의 초기 신호를 포착할 수 있는 능력을 보여줍니다. 다음은 표의 주요 항목과 설명입니다.

### 주요 항목:
- **Spatial Resolution (공간 해상도)**: 30m, 70m / 500m
- **Spatial Coverage (공간 범위)**: CONUS (미국 본토) / 전 세계(Global)
- **Temporal Resolution (시간 해상도)**: 특정 지역은 1~7일 간격으로 관측 / 며칠 단위로 관측
- **Temporal Coverage (시간 범위)**: 2018년부터 현재까지 / 2001년부터 현재까지
- **Spectral Resolution (스펙트럼 해상도)**: 
  - 5개의 밴드 (8µm ~ 12.5µm, 추가적으로 1.6µm 밴드는 위치 추적 및 구름 감지를 위해 사용)
  - 36개의 밴드 (0.4µm ~ 14.4µm)
- **Satellite/Platform (위성/플랫폼)**: 국제우주정거장(International Space Station) / Terra 위성
- **Sensor(s)/Model Name (센서/모델 이름)**: ECOSTRESS / MODIS
- **Observation or Model (관측 또는 모델)**: 모델링 기반 데이터
- **File Format (파일 형식)**: GeoTIFF / HDF-EOS

이 데이터는 전 세계적인 증발 스트레스 및 토양 수분 고갈을 감시하고 분석하는 데 사용됩니다.

# 증발산(ET, 
)
**증발산(ET, Evapotranspiration)**에 대한 데이터를 요약한 표가 있습니다. 증발산은 지표면에서의 증발과 식물의 증산을 합친 개념으로, 물 가용성, 가뭄 상태, 작물 생산 등을 평가하는 데 유용합니다. 이를 통해 작물의 물 사용량을 예측하고, 농업용수 관리에 도움이 됩니다.

### 주요 항목:

- **Spatial Resolution (공간 해상도)**: 30m, 70m / 500m / 0.01° ~ 0.25°
- **Spatial Coverage (공간 범위)**: 전 세계(Global) 및 CONUS (미국 본토)
- **Temporal Resolution (시간 해상도)**: 약 1~7일 / 1~2일 / 16일 / 시간별, 3시간별, 일별, 월별
- **Temporal Coverage (시간 범위)**: 2018년부터 현재 / 2000년부터 현재 / 1982년부터 현재 / 1984년부터 현재
- **Spectral Resolution (스펙트럼 해상도)**: 
  - 5개의 밴드 (8 µm ~ 12.5 µm, 위치 추적 및 구름 감지를 위한 1.6 µm 추가 밴드 포함)
  - 36개의 밴드 (0.4 µm ~ 14.4 µm)
  - OLI/TIRS: 9개의 밴드 (0.43 µm ~ 1.38 µm), TIRS: 2개의 밴드 (10.6 µm ~ 12.51 µm)
  - ETM+: 8개의 밴드 (0.45 µm ~ 12.5 µm), TM: 7개의 밴드 (0.45 µm ~ 2.35 µm)
- **Satellite/Platform (위성/플랫폼)**: 국제우주정거장(ISS) / Terra / Aqua / Landsat 4, 5, 7, 8, 9
- **Sensor(s)/Model Name (센서/모델 이름)**: ECOSTRESS / MODIS / OLI/TIRS, ETM+, TM / LDAS
- **Observation or Model (관측 또는 모델)**: 관측 / 관측 및 모델링 / 모델링
- **File Format (파일 형식)**: GeoTIFF / HDF5 / HDF-EOS / netCDF

이 데이터는 전 세계적으로 ET 추정치를 제공하여, 농업과 물 관리 계획에 중요한 자료로 사용됩니다.

### 주요 변수 설명:

1. **`time_bnds`**:  
   - 시간 경계 값. 시간 변수와 관련된 경계(boundaries)를 나타냅니다.

2. **`Evap_tavg`**:  
   - **평균 증발산(Evapotranspiration)**.  
   - 지면에서 증발되고 식물에서 증발하는 물의 양을 나타내는 변수로, 물 순환과 에너지 흐름에서 중요한 역할을 합니다. 이 데이터셋에서의 증발산량은 주어진 시간 동안 평균된 값입니다.  
   - 이 데이터가 **Evapotranspiration**(증발산)과 관련된 데이터입니다.

3. **`LWdown_f_tavg`**:  
   - **평균 하향 장파복사(LWdown)**.  
   - 대기에서 지표면으로 내려오는 장파복사(thermal radiation)의 평균 값입니다.

4. **`Lwnet_tavg`**:  
   - **평균 순 장파복사(Lwnet)**.  
   - 장파복사의 순 값, 즉 지표에서 방출된 장파복사에서 대기로부터 받은 장파복사를 뺀 값입니다.

5. **`Psurf_f_tavg`**:  
   - **평균 지표면 기압(Psurf)**.  
   - 지표면에서의 대기 압력 평균값입니다.

6. **`Qair_f_tavg`**:  
   - **평균 절대 습도(Qair)**.  
   - 대기 중의 수증기 양(절대 습도)의 평균 값입니다.

7. **`Qg_tavg`**:  
   - **평균 지면 열 흐름(Qg)**.  
   - 지면에서 발생하는 열 흐름의 평균값입니다.

8. **`Qh_tavg`**:  
   - **평균 감열(Qh)**.  
   - 감열(대기와 지표 간의 열 전달) 흐름의 평균 값입니다.

9. **`Qle_tavg`**:  
   - **평균 잠열(Qle)**.  
   - 대기와 지면 간의 수증기 교환에 따른 열 흐름을 나타내는 평균 잠열 값입니다.

10. **`Qs_tavg`**:  
   - **평균 지표 흐름(Qs)**.  
   - 지표면에서의 물 흐름(표면 유출)의 평균값입니다.

11. **`Qsb_tavg`**:  
   - **평균 하부 지표 흐름(Qsb)**.  
   - 지하수 흐름과 관련된 평균 하부 지표수 유출량입니다.

12. **`RadT_tavg`**:  
   - **평균 복사 온도(RadT)**.  
   - 복사에 의해 전달된 에너지를 나타내는 평균 온도입니다.

13. **`Rainf_f_tavg`**:  
   - **평균 강수량(Rainf)**.  
   - 평균 강수량 또는 강수 흐름의 양을 나타냅니다.

14. **`SWE_inst`**:  
   - **즉시 눈물(Snow Water Equivalent, SWE)**.  
   - 눈이 물로 변환된 후의 양(물 동등량)을 나타냅니다.

15. **`SWdown_f_tavg`**:  
   - **평균 하향 단파복사(SWdown)**.  
   - 태양으로부터 지표로 향하는 단파복사(일사량)의 평균 값입니다.

16. **`SnowCover_inst`**:  
   - **즉시 눈 덮임(SnowCover)**.  
   - 즉시 눈으로 덮인 지표면 비율을 나타냅니다.

17. **`SnowDepth_inst`**:  
   - **즉시 눈 깊이(SnowDepth)**.  
   - 지표면에 쌓인 눈의 깊이를 나타냅니다.

18. **`Snowf_tavg`**:  
   - **평균 눈 강수(Snowfall)**.  
   - 눈의 강수량을 나타냅니다.

19. **`Swnet_tavg`**:  
   - **평균 순 단파복사(Swnet)**.  
   - 지표에서 흡수된 단파복사에서 반사된 단파복사를 뺀 순 단파복사 값입니다.

20. **`Tair_f_tavg`**:  
   - **평균 대기 온도(Tair)**.  
   - 주어진 높이에서의 대기 온도의 평균 값입니다.

21. **`Wind_f_tavg`**:  
   - **평균 바람 속도(Wind)**.

22. **`SoilMoi00_10cm_tavg`**:

### **Evapotranspiration 데이터**:
`Evap_tavg`가 바로 **Evapotranspiration(증발산)** 데이터를 나타내며, 이는 증발과 식물 증산량을 포함한 수증기 흐름을 나타냅니다. Evapotranspiration은 농업, 수자원 관리, 기후 연구에서 중요한 역할을 하며, 이 데이터는 해당 지역의 물 손실량을 예측하거나 측정하는 데 사용될 수 있습니다.

이 데이터를 분석하고 시각화하는 데 `Evap_tavg` 변수를 활용할 수 있습니다.

```python
import matplotlib.pyplot as plt

# Evap_tavg 변수 추출
evap_data = ds['Evap_tavg']

# 특정 시간에 대한 데이터 선택 (예: 첫 번째 time step)
time_index = 0  # 첫 번째 시간
evap_at_time = evap_data.isel(time=time_index)

# Evapotranspiration 데이터 시각화
plt.figure(figsize=(10, 6))
evap_at_time.plot(cmap='viridis')  # colormap은 취향에 따라 변경 가능
plt.title(f'Evapotranspiration at time index {time_index}')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import geopandas as gpd
import xarray as xr
from shapely.geometry import mapping

# GeoJSON 파일 로드
geojson_path = './egypt.geojson'
egypt_boundary = gpd.read_file(geojson_path)

# 경계를 시각화(선택적)
egypt_boundary.plot()
plt.title('Egypt Boundary')
plt.show()
```

```python
import rioxarray
from rioxarray import merge
import matplotlib.pyplot as plt

# xarray 객체에 좌표 시스템 설정
evap_data = ds['Evap_tavg']
evap_data = evap_data.rio.write_crs("EPSG:4326")  # EPSG:4326은 WGS84 좌표계입니다

# 이집트 경계 다각형 추출
egypt_polygon = egypt_boundary.geometry.unary_union  # 경계가 여러 개일 경우 병합

# 이집트 경계 내의 Evapotranspiration 데이터만 추출
clipped_evap = evap_data.rio.clip([mapping(egypt_polygon)], evap_data.rio.crs, drop=True)

# 시각화
plt.figure(figsize=(10, 6))
clipped_evap.isel(time=0).plot(cmap='viridis')
plt.title('Evapotranspiration in Egypt (First Time Step)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# 작년과 올해 비교 

```python
import xarray as xr
import matplotlib.pyplot as plt

# 두 파일 경로 지정
file_path_2024 = './FLDAS_NOAH01_C_GL_M.A202408.001.nc'
file_path_2023 = './FLDAS_NOAH01_C_GL_M.A202308.001.nc'

# 두 파일의 NetCDF 파일 열기
ds_2024 = xr.open_dataset(file_path_2024)
ds_2023 = xr.open_dataset(file_path_2023)

# Evap_tavg 변수 추출
evap_2024 = ds_2024['Evap_tavg']
evap_2023 = ds_2023['Evap_tavg']

# 특정 시간에 대한 데이터 선택 (첫 번째 time step)
time_index = 0  # 첫 번째 시간

# 2024년 8월 데이터
evap_at_time_2024 = evap_2024.isel(time=time_index)

# 2023년 8월 데이터
evap_at_time_2023 = evap_2023.isel(time=time_index)

# 두 데이터를 비교하는 시각화
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# 2024년 8월 시각화
evap_at_time_2024.plot(ax=axes[0], cmap='viridis')
axes[0].set_title(f'Evapotranspiration - 2024 (Time index {time_index})')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')

# 2023년 8월 시각화
evap_at_time_2023.plot(ax=axes[1], cmap='viridis')
axes[1].set_title(f'Evapotranspiration - 2023 (Time index {time_index})')
axes[1].set_xlabel('Longitude')
axes[1].set_ylabel('Latitude')

plt.tight_layout()
plt.show()

# 두 데이터의 차이를 계산하고 시각화
evap_diff = evap_at_time_2024 - evap_at_time_2023

plt.figure(figsize=(10, 6))
evap_diff.plot(cmap='RdBu_r')  # 차이값 시각화 (양수와 음수 차이를 표현하는 색상)
plt.title('Difference in Evapotranspiration (2024 - 2023)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import rioxarray
from rioxarray import merge
import geopandas as gpd
from shapely.geometry import mapping
import matplotlib.pyplot as plt
import xarray as xr

# GeoJSON 파일 로드
geojson_path = './egypt.geojson'
egypt_boundary = gpd.read_file(geojson_path)

# 두 파일 경로 지정
file_path_2024 = './FLDAS_NOAH01_C_GL_M.A202408.001.nc'
file_path_2023 = './FLDAS_NOAH01_C_GL_M.A202308.001.nc'

# 두 파일의 NetCDF 파일 열기
ds_2024 = xr.open_dataset(file_path_2024)
ds_2023 = xr.open_dataset(file_path_2023)

# Evap_tavg 변수 추출
evap_2024 = ds_2024['Evap_tavg']
evap_2023 = ds_2023['Evap_tavg']

# xarray 객체에 좌표 시스템 설정 (EPSG:4326은 WGS84 좌표계)
evap_2024 = evap_2024.rio.write_crs("EPSG:4326")
evap_2023 = evap_2023.rio.write_crs("EPSG:4326")

# 이집트 경계 다각형 추출
egypt_polygon = egypt_boundary.geometry.unary_union

# 이집트 경계 내의 Evapotranspiration 데이터만 추출 (첫 번째 시간 스텝)
evap_2024_clipped = evap_2024.isel(time=0).rio.clip([mapping(egypt_polygon)], evap_2024.rio.crs, drop=True)
evap_2023_clipped = evap_2023.isel(time=0).rio.clip([mapping(egypt_polygon)], evap_2023.rio.crs, drop=True)

# 시각화: 2024년과 2023년 증발산 비교
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# 2024년 데이터 시각화
evap_2024_clipped.plot(ax=axes[0], cmap='viridis')
axes[0].set_title('Evapotranspiration in Egypt - 2024')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')

# 2023년 데이터 시각화
evap_2023_clipped.plot(ax=axes[1], cmap='viridis')
axes[1].set_title('Evapotranspiration in Egypt - 2023')
axes[1].set_xlabel('Longitude')
axes[1].set_ylabel('Latitude')

# 두 데이터 차이 계산 (2024 - 2023)
evap_diff = evap_2024_clipped - evap_2023_clipped

# 차이 시각화
evap_diff.plot(ax=axes[2], cmap='RdBu_r')
axes[2].set_title('Difference in Evapotranspiration (2024 - 2023)')
axes[2].set_xlabel('Longitude')
axes[2].set_ylabel('Latitude')

plt.tight_layout()
plt.show()
```

증발산량(Evapotranspiration)은 지표면에서 대기로 증발하는 물의 양과 식물에서 방출되는 증산을 모두 포함하는 개념입니다. 따라서 증발산량이 높은지 낮은지는 **농업 환경**, **물 자원 가용성**, 그리고 **작물 유형**에 따라 다르게 평가될 수 있습니다.

### **증발산량이 높고 낮은 것이 의미하는 바**:
1. **증발산량이 높은 경우**:
   - **장점**:
     - 식물의 증산이 활발하다는 것은 식물의 성장에 필요한 물이 충분히 공급되고 있으며, 광합성도 잘 이루어지고 있음을 의미합니다.
   - **단점**:
     - 그러나 너무 많은 증발산은 토양 수분의 손실을 가속화할 수 있습니다. 이는 특히 건조하거나 물이 부족한 지역에서는 농업에 부정적인 영향을 미칠 수 있습니다. 충분한 관개가 필요한 상황입니다.
   
2. **증발산량이 낮은 경우**:
   - **장점**:
     - 토양의 수분이 잘 보존되고, 작물에게 필요한 수분이 적게 손실될 수 있습니다. 이는 건조한 환경에서 물 자원을 절약할 수 있다는 점에서 긍정적일 수 있습니다.
   - **단점**:
     - 그러나 낮은 증발산량은 식물이 충분한 물을 흡수하지 못하고, 성장에 필요한 수분을 받지 못할 수도 있다는 뜻입니다. 이는 농작물의 성장을 저해할 수 있습니다.

### **2023년과 2024년의 비교 - 어떤 해가 농업에 유리했는가?**
- **2023년**:
  - 증발산량이 상대적으로 낮았습니다. 이 말은 **토양의 수분이 더 오래 보존**되었을 가능성이 크다는 것을 의미합니다. 그러나 식물의 성장은 그만큼 활발하지 않았을 수 있습니다.
  
- **2024년**:
  - 증발산량이 높았으며, 이는 **식물의 증산이 활발**했음을 나타냅니다. 하지만 그만큼 토양의 수분 손실도 컸을 가능성이 있으며, **충분한 관개**가 필요한 상황이었을 수 있습니다.

### **농업에 유리한 해는?**
- **농업이 잘 되기 위해서는** 적절한 증발산량과 충분한 수분 공급이 필요합니다. 증발산량이 너무 낮으면 작물이 성장하는 데 필요한 물을 충분히 공급받지 못하고, 너무 높으면 물 자원이 빨리 고갈될 수 있습니다.
  
- **2024년**은 증발산량이 더 높았으므로, 그 해에 물 자원이 충분히 공급되었다면 **작물의 성장이 활발했을 가능성**이 큽니다. 그러나 **관개가 충분하지 않은 경우** 토양의 수분 손실이 커서 **농업에 부정적인 영향**이 있었을 수도 있습니다.

- **2023년**은 증발산량이 낮았으므로, **물 자원 관리가 덜 필요했을 수 있지만**, 작물의 성장도 그만큼 더디었을 가능성이 있습니다.

### **결론**:
- **관개가 충분히 이루어진 경우**: 2024년이 더 유리했을 수 있습니다. 식물의 성장이 활발했을 가능성이 크기 때문입니다.
- **관개가 부족하거나 물 자원이 부족한 경우**: 2023년이 더 유리했을 수 있습니다. 토양 수분이 덜 손실되었기 때문입니다.

따라서 증발산량이 높고 낮은 것은 **물 자원 관리 상황**에 따라 다르게 평가될 수 있습니다. 물이 충분히 공급되는 환경에서는 **2024년**이 더 농업에 적합했을 수 있습니다.
