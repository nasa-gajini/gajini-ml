# MODIS/Terra Vegetation Indices 16-Day L3 Global 1km SIN Grid V061
* DOI 10.5067/MODIS/MOD13A2.061
* https://lpdaac.usgs.gov/products/mod13a2v061/
* How to read hdf file 
    * https://hdfeos.org/software/pyhdf.php   

이 설명은 **MOD13A2**라는 Terra 위성의 **MODIS (Moderate Resolution Imaging Spectroradiometer)** 데이터를 설명하고 있으며, 1km 해상도의 **16일 Vegetation Index (VI)** 제품에 포함된 변수들과 그 의미를 제공하고 있습니다. MOD13A2는 식생 지수를 제공하는 위성 데이터 제품으로, 이 데이터는 주로 **NDVI**와 **EVI**라는 두 가지 주요 식생 지수를 기반으로 합니다.

### 주요 내용 설명

1. **NDVI (Normalized Difference Vegetation Index)**:
   - **NDVI**는 기존 NOAA-AVHRR에서 사용하는 NDVI와의 연속성을 유지하는 식생 지수입니다.
   - 이 지수는 식물의 상태를 측정하는 데 사용되며, 값이 클수록 건강한 식생을 나타냅니다.
   - 데이터는 16일 동안 수집된 값 중, 구름이 적고 관측 각도가 좋은 조건에서 최고의 NDVI 값을 선택하여 제공됩니다.

2. **EVI (Enhanced Vegetation Index)**:
   - **EVI**는 NDVI보다 높은 생물량(높은 식물 덮개가 있는 지역)을 더 정확하게 감지할 수 있도록 개선된 지수입니다.
   - NDVI와 유사하게, 16일 동안 수집된 값 중에서 최적의 값이 선택됩니다.

### 데이터 세트 설명 (Layers / Variables)

각 변수(레이어)에 대한 설명은 아래와 같습니다:

1. **1 km 16 days NDVI**:
   - 16일 동안의 1km 해상도 NDVI 데이터입니다.
   - **값 범위**는 -2000에서 10000까지이며, **스케일 팩터**는 0.0001입니다. 실제 값을 계산하려면 원본 값에 스케일 팩터를 곱해야 합니다.
   - 예를 들어, NDVI 값이 8000이면 실제 값은 0.8입니다 (8000 * 0.0001 = 0.8).

2. **1 km 16 days EVI**:
   - 16일 동안의 1km 해상도 EVI 데이터입니다.
   - NDVI와 마찬가지로 **스케일 팩터**는 0.0001이며, 값 범위는 -2000에서 10000까지입니다.

3. **1 km 16 days VI Quality**:
   - NDVI와 EVI에 대한 품질 지표(품질 보증 정보)를 포함하는 비트 필드 데이터입니다.
   - 이 변수는 데이터를 필터링하거나 품질을 확인할 때 사용됩니다.

4. **1 km 16 days Red Reflectance**:
   - 지표의 **적색 밴드 반사율**을 나타냅니다 (Band 1, 적색 대역).
   - 스케일 팩터는 0.0001이며, 값 범위는 0에서 10000입니다.

5. **1 km 16 days NIR Reflectance**:
   - **근적외선(NIR)** 대역의 반사율을 나타냅니다 (Band 2).
   - NIR 반사율은 식생 상태를 분석하는 데 중요한 정보입니다. 식물이 건강할수록 더 많은 근적외선을 반사합니다.

6. **1 km 16 days Blue Reflectance**:
   - **청색 밴드 반사율** (Band 3).

7. **1 km 16 days MIR Reflectance**:
   - **중적외선(MIR)** 대역의 반사율 (Band 7).

8. **1 km 16 days View Zenith Angle**:
   - **위성 관측 시의 시야 천정각**을 나타냅니다.
   - 값 범위는 0에서 18000이며, 스케일 팩터는 0.01입니다. 실제 값을 계산하려면 원본 값에 0.01을 곱해야 합니다.

9. **1 km 16 days Sun Zenith Angle**:
   - **태양 천정각**을 나타냅니다.
   - 값 범위와 스케일 팩터는 시야 천정각과 동일합니다.

10. **1 km 16 days Relative Azimuth Angle**:
    - **상대 방위각**을 나타냅니다.
    - 값 범위는 -18000에서 18000이며, 스케일 팩터는 0.01입니다.

11. **1 km 16 days Composite Day of the Year**:
    - 1년 중 어느 날에 해당 VI(식생 지수) 값이 생성되었는지를 나타내는 줄리안 날짜입니다.
    - 값 범위는 1에서 366까지입니다.

12. **1 km 16 days Pixel Reliability**:
    - 각 VI 픽셀의 **신뢰도**를 나타내는 값입니다.
    - 이 값은 데이터의 신뢰도를 나타내며, 값이 클수록 신뢰도가 낮습니다.

### 요약:
- **NDVI**와 **EVI**는 식생 상태를 분석하는 데 사용되는 주요 지표입니다.
- **반사율** 밴드는 식생 지수 계산을 위한 기초 데이터로, 적색, 근적외선, 청색, 중적외선 대역의 반사율을 제공합니다.
- **품질 지표와 각도**(천정각, 방위각)는 데이터의 정확도를 평가하는 데 사용됩니다.
- **줄리안 날짜**와 **신뢰도**는 각 데이터의 시간적 정보를 제공하며, 신뢰도가 낮은 데이터를 필터링할 수 있습니다.

이 변수들을 활용해 식생 지수를 분석할 때 데이터를 필터링하거나 대기 보정, 각도 보정 등을 수행할 수 있습니다.

```python
import numpy as np
from pyhdf.SD import SD, SDC

# HDF 파일 경로들
hdf_file_paths = [
    "./data/Vegetation_NDVI/MOD13A2.A2024257.h20v05.061.2024275121224.hdf",  # 왼쪽 위
    "./data/Vegetation_NDVI/MOD13A2.A2024257.h20v06.061.2024275115107.hdf",  # 왼쪽 아래
    "./data/Vegetation_NDVI/MOD13A2.A2024257.h21v06.061.2024275115124.hdf",  # 오른쪽 아래
    "./data/Vegetation_NDVI/MOD13A2.A2024257.h21v05.061.2024275120200.hdf"   # 오른쪽 위
]

# 결합할 데이터셋 리스트 (모든 dataset_name)
dataset_names = [
    '1 km 16 days Vegetation_NDVI',
    '1 km 16 days EVI',
    '1 km 16 days VI Quality',
    '1 km 16 days red reflectance',
    '1 km 16 days NIR reflectance',
    '1 km 16 days blue reflectance',
    '1 km 16 days MIR reflectance',
    '1 km 16 days view zenith angle',
    '1 km 16 days sun zenith angle',
    '1 km 16 days relative azimuth angle',
    '1 km 16 days composite day of the year',
    '1 km 16 days pixel reliability'
]

# 타일 크기
tile_size = (1200, 1200)

# 모든 데이터셋을 결합
for dataset_name in dataset_names:
    # 전체 배열 크기 (2400x2400)
    full_size = (2400, 2400)
    combined_array = np.zeros(full_size, dtype=np.float32)
    
    # 각 타일의 위치에 맞게 데이터를 읽어서 결합
    for idx, hdf_file_path in enumerate(hdf_file_paths):
        # HDF 파일 열기
        hdf = SD(hdf_file_path, SDC.READ)

        # 데이터셋 읽기
        dataset = hdf.select(dataset_name)
        data = dataset[:]

        # 각 타일의 위치에 맞게 배열에 삽입
        if idx == 0:  # 왼쪽 위
            combined_array[0:tile_size[0], 0:tile_size[1]] = data
        elif idx == 1:  # 왼쪽 아래
            combined_array[tile_size[0]:full_size[0], 0:tile_size[1]] = data
        elif idx == 2:  # 오른쪽 아래
            combined_array[tile_size[0]:full_size[0], tile_size[1]:full_size[1]] = data
        elif idx == 3:  # 오른쪽 위
            combined_array[0:tile_size[0], tile_size[1]:full_size[1]] = data

    # 각 데이터셋에 대한 결합 결과 출력 (배열 크기 확인)
    print(f"{dataset_name} 결합 완료: {combined_array.shape}")
```

```python
import numpy as np

# 모든 데이터셋을 결합
for dataset_name in dataset_names:
    # 전체 배열 크기 (2400x2400)
    full_size = (2400, 2400)
    combined_array = np.zeros(full_size, dtype=np.float32)
    
    # 각 타일의 위치에 맞게 데이터를 읽어서 결합
    for idx, hdf_file_path in enumerate(hdf_file_paths):
        # HDF 파일 열기
        hdf = SD(hdf_file_path, SDC.READ)

        # 데이터셋 읽기
        dataset = hdf.select(dataset_name)
        data = dataset[:]

        # 각 타일의 위치에 맞게 배열에 삽입
        if idx == 0:  # 왼쪽 위
            combined_array[0:tile_size[0], 0:tile_size[1]] = data
        elif idx == 1:  # 왼쪽 아래
            combined_array[tile_size[0]:full_size[0], 0:tile_size[1]] = data
        elif idx == 2:  # 오른쪽 아래
            combined_array[tile_size[0]:full_size[0], tile_size[1]:full_size[1]] = data
        elif idx == 3:  # 오른쪽 위
            combined_array[0:tile_size[0], tile_size[1]:full_size[1]] = data

    # 결합된 결과를 파일로 저장 (예: NumPy 파일로 저장)
    output_filename = f"{dataset_name.replace(' ', '_')}_combined.npy"
    np.save(output_filename, combined_array)
    
    # 결합된 파일 경로 출력
    print(f"{dataset_name} 결합 완료 및 저장: {output_filename}")
```

# Web에 전달

```python
 import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import folium
import os

# NumPy 파일 로드 (이미 결합된 Vegetation_NDVI 데이터)
ndvi_combined = np.load('1_km_16_days_NDVI_combined.npy')

# Vegetation_NDVI 시각화를 위한 컬러맵 설정
cmap = plt.get_cmap('RdYlGn')
norm = mcolors.Normalize(vmin=-2000, vmax=10000)

# Vegetation_NDVI 데이터를 0~255 범위의 값으로 변환하여 이미지 생성
ndvi_image = np.uint8(cmap(norm(ndvi_combined)) * 255)

# 이미지 파일로 저장
output_image_path = "ndvi_image.png"
plt.imsave(output_image_path, ndvi_image)

# 이미지 저장 위치 확인
print(f"Vegetation_NDVI 이미지 저장 완료: {output_image_path}")
```

```python
import folium
from folium import plugins

# Leaflet에서 사용할 타일 범위 (예시)
min_lat, max_lat = 20.0, 41.177386
min_lon, max_lon = -60.9454, -30.0

# Folium 지도 생성 (중심점 및 줌 설정)
m = folium.Map(location=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2], zoom_start=5)

# Vegetation_NDVI 이미지 타일 추가
folium.raster_layers.ImageOverlay(
    image=output_image_path,
    bounds=[[min_lat, min_lon], [max_lat, max_lon]],
    opacity=0.7,
    interactive=True,
).add_to(m)

# 지도 저장
m.save("ndvi_map.html")
print("지도가 생성되어 ndvi_map.html로 저장되었습니다.")
```
