```python
import xarray as xr
import requests
import datetime as dt
import pprint
import netCDF4 as nc4
```

```python
# CMR API base url
cmrurl='https://cmr.earthdata.nasa.gov/search/' # define the base url of NASA's CMR API as the variable `cmrurl`
doi = '10.5067/MODIS/MCD19A2.061'                   # Daymet V4R1 DOI - define the DOI as the variable `daymet_doi`
doisearch = cmrurl + 'collections.json?doi=' + doi
print(doisearch)
```

```python
concept_id = requests.get(doisearch).json()['feed']['entry'][0]['id']
print(concept_id)
```

```python
start_date = dt.datetime(2024, 9, 29) # specify your own start date
end_date = dt.datetime(2024, 9, 30)  # specify your end start date
print(start_date,end_date,sep='\n')
```

```python
dt_format = '%Y-%m-%dT%H:%M:%SZ' # format requirement for datetime search
temporal_str = start_date.strftime(dt_format) + ',' + end_date.strftime(dt_format)
print(temporal_str)
```

```python
cmr_url = 'https://cmr.earthdata.nasa.gov/search/granules'
```

```python
cmr_response = requests.get(cmr_url, 
                            params={
                                'concept_id': concept_id,
                                'temporal': temporal_str,
                                'page_size': 200,
                                },
                            headers={
                                'Accept': 'application/json'
                                }
                            )

cmr_response.json()
```

```python
cmr_response.json()['feed']['entry']
```

```python
type(cmr_response.json()['feed']['entry'])
```

# S3  

```python
entries = cmr_response.json()['feed']['entry']
s3_links = []

for entry in entries:
    for link in entry['links']:  # 'links'      
        if 's3' in link['href']:  # S3 URL 
            s3_links.append(link['href'])

#  
for s3_link in s3_links:
    print(s3_link)
```

#     

```python
metadata_links = []

for entry in entries:
    for link in entry['links']:
        if 'metadata' in link['rel']:  #    
            metadata_links.append(link['href'])

#  
for metadata_link in metadata_links:
    print(metadata_link)
```
