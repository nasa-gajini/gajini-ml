```python
import h5py

# 모든 변수 이름을 출력하는 함수
def print_hdf5_variable_names(file_path):
    # HDF5 파일 열기
    with h5py.File(file_path, 'r') as file:
        # 파일의 모든 key (그룹 및 데이터셋 이름) 출력
        def print_name(name, obj):
            if isinstance(obj, h5py.Dataset):
                print(name)
        
        # 모든 그룹 및 데이터셋 이름 출력
        file.visititems(print_name)

# Main 실행 부분
if __name__ == '__main__':
    # HDF5 파일 경로
    file_path = '../data/SMAP/SMAP_L3_SM_P_E_20240801_R19240_002.h5'
    
    # 변수 이름 출력
    print_hdf5_variable_names(file_path)
```
